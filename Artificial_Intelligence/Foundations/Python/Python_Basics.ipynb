## 10. Exception Handling

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print(f"Error: {e}")
        return None
    except TypeError as e:
        print(f"Type error: {e}")
        return None
    else:
        print("Division succeeded.")
        return result
    finally:
        print("Cleanup done.")

safe_divide(10, 2)
safe_divide(10, 0)
```

```python
# Custom exception
class NegativeValueError(Exception):
    pass

def sqrt_safe(x):
    if x < 0:
        raise NegativeValueError(f"Cannot compute sqrt of {x}")
    return x ** 0.5

try:
    print(sqrt_safe(9))
    print(sqrt_safe(-1))
except NegativeValueError as e:
    print(e)
```

## 17. Closures and Decorators

```python
# Closure
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

double = make_multiplier(2)
print(double(7))

# Decorator
from functools import wraps

def log(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"→ Calling {func.__name__}({args}, {kwargs})")
        result = func(*args, **kwargs)
        print(f"← Returned: {result}")
        return result
    return wrapper

@log
def add(a, b):
    return a + b

add(3, 4)
```

## 18. Generators

```python
# Generator function
def count_up(n):
    for i in range(n):
        yield i

gen = count_up(5)
print(next(gen))
print(next(gen))
print(list(count_up(5)))

# Generator expression (lazy evaluation)
squares_gen = (x**2 for x in range(10))
print(list(squares_gen))

# Infinite generator (take with care)
def naturals():
    n = 0
    while True:
        yield n
        n += 1

import itertools
first_10 = list(itertools.islice(naturals(), 10))
print(first_10)
```

## 19. Classes and Objects

```python
class Animal:
    kingdom = "Animalia"   # Class variable

    def __init__(self, name: str, sound: str):
        self.name  = name   # Instance variable
        self.sound = sound

    def speak(self) -> str:
        return f"{self.name} says {self.sound}"

    def __repr__(self):
        return f"Animal({self.name!r})"

    def __str__(self):
        return self.name


dog = Animal("Dog", "Woof")
print(dog.speak())
print(repr(dog))
print(str(dog))
print(Animal.kingdom)
```

### Inheritance

```python
class Dog(Animal):
    def __init__(self, name: str, breed: str):
        super().__init__(name, "Woof")
        self.breed = breed

    def fetch(self) -> str:
        return f"{self.name} fetches the ball!"

    def speak(self) -> str:      # Override
        return f"{self.name} barks loudly!"


lab = Dog("Rex", "Labrador")
print(lab.speak())
print(lab.fetch())
print(isinstance(lab, Dog))
print(isinstance(lab, Animal))
```

### Special (Dunder) Methods

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __len__(self):
        return 2

    def __getitem__(self, index):
        return (self.x, self.y)[index]

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)
print(len(v1))
print(v1[0])
print(v1 == Vector(1, 2))
```

### Properties and Class/Static Methods

```python
import math

class Circle:
    def __init__(self, radius: float):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Radius must be non-negative.")
        self._radius = value

    @property
    def area(self):
        return math.pi * self._radius ** 2

    @classmethod
    def unit(cls):
        return cls(1.0)

    @staticmethod
    def is_positive(n):
        return n > 0


c = Circle(5)
c.radius = 10
print(f"Radius: {c.radius}, Area: {c.area:.2f}")
print(Circle.unit())
print(Circle.is_positive(-1))
```

### Abstract Classes

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self) -> float: pass

    @abstractmethod
    def perimeter(self) -> float: pass


class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)


r = Rectangle(4, 6)
print(f"Area: {r.area()}, Perimeter: {r.perimeter()}")
```

## 20. Dataclasses

```python
from dataclasses import dataclass, field
from typing import List
import statistics

@dataclass
class Student:
    name: str
    grade: int
    scores: List[float] = field(default_factory=list)

    def add_grade(self, score: float) -> None:
        if not (0 <= score <= 100):
            raise ValueError(f"Score {score} out of range.")
        self.scores.append(score)

    def average(self) -> float:
        return statistics.mean(self.scores) if self.scores else 0.0

    def letter_grade(self) -> str:
        avg = self.average()
        if avg >= 90: return "A"
        elif avg >= 80: return "B"
        elif avg >= 70: return "C"
        elif avg >= 60: return "D"
        return "F"


s = Student("Alice", 10)
for g in [92, 88, 95, 79]:
    s.add_grade(g)

print(s)
print(f"Average: {s.average():.2f}, Grade: {s.letter_grade()}")
```

## 23. Context Managers

```python
from contextlib import contextmanager
import time

@contextmanager
def timer(label=""):
    start = time.perf_counter()
    yield
    elapsed = time.perf_counter() - start
    print(f"{label} Elapsed: {elapsed:.6f}s")

with timer("Sum 1M:"):
    total = sum(range(1_000_000))
    
print(f"Total: {total}")
```

## 24. Regular Expressions

```python
import re

text = "Contact us at support@example.com or info@domain.org"

# Search (first match)
match = re.search(r"\w+@\w+\.\w+", text)
if match:
    print("First:", match.group())

# Find all
emails = re.findall(r"\w+@\w+\.\w+", text)
print("All:", emails)

# Substitute
clean = re.sub(r"\d+", "#", "Item 42 costs 100 units")
print(clean)

# Groups
pattern = re.compile(r"(\w+)@(\w+)\.(\w+)")
for m in pattern.finditer(text):
    print(f"User: {m.group(1)}, Domain: {m.group(2)}.{m.group(3)}")
```

## 25. Itertools and Functools

```python
import itertools
from functools import reduce, lru_cache

# Combinations & permutations
print(list(itertools.combinations([1,2,3], 2)))
print(list(itertools.permutations([1,2,3], 2)))

# Chain
print(list(itertools.chain([1,2], [3,4], [5])))

# Accumulate (running total)
print(list(itertools.accumulate([1,2,3,4])))

# reduce
print(reduce(lambda a, b: a * b, [1,2,3,4,5]))   # 120

# LRU cache (memoization)
@lru_cache(maxsize=None)
def fib(n):
    if n < 2: return n
    return fib(n-1) + fib(n-2)

print([fib(i) for i in range(10)])
```

## 26. Type Hints (PEP 484)

```python
from typing import List, Dict, Optional, Tuple, Union

def greet(name: str) -> str:
    return f"Hello, {name}"

def process(data: List[int]) -> Dict[str, int]:
    return {"sum": sum(data), "count": len(data)}

def find(name: str) -> Optional[str]:
    return name if name else None

def identity(x: Union[int, float]) -> Union[int, float]:
    return x

print(greet("Alice"))
print(process([1, 2, 3, 4, 5]))
print(find(""))
print(identity(3.14))
```

## 💻 Full Example — Student Grade Tracker

A small program combining dataclasses, type hints, comprehensions, exception handling, and built-ins.

```python
from dataclasses import dataclass, field
from typing import List, Dict
import statistics

@dataclass
class Student:
    name: str
    grades: List[float] = field(default_factory=list)

    def add_grade(self, grade: float) -> None:
        if not (0 <= grade <= 100):
            raise ValueError(f"Grade {grade} is out of range.")
        self.grades.append(grade)

    def average(self) -> float:
        return statistics.mean(self.grades) if self.grades else 0.0

    def letter_grade(self) -> str:
        avg = self.average()
        if avg >= 90: return "A"
        elif avg >= 80: return "B"
        elif avg >= 70: return "C"
        elif avg >= 60: return "D"
        return "F"


def class_report(students: List[Student]) -> Dict[str, float]:
    return {s.name: s.average() for s in students}


# ── Run ────────────────────────────────────────────────────────
alice = Student("Alice")
bob   = Student("Bob")

for g in [92, 88, 95, 79]: alice.add_grade(g)
for g in [70, 65, 72, 68]: bob.add_grade(g)

report = class_report([alice, bob])

print("=== Class Report ===")
for name, avg in report.items():
    obj = alice if name == "Alice" else bob
    print(f"  {name:<10} avg={avg:5.2f}  grade={obj.letter_grade()}")

top = max(report, key=report.get)
print(f"\nTop student: {top} ({report[top]:.2f})")
```
